```python
from langchain_google_genai import GoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
prompt = PromptTemplate(
    input_variables = ['question'],
    template = """ Explain '{question}' in simple terms with an example:

    Format:
    -keep it under 100 words
    -use an analogy
    -End with a practical example

    question: {question}
    explanation: """
)
filled_prompt = prompt.format(question = 'AI agent')
print(filled_prompt)
```

```python
#first give the prompt,
#then connect to llm and get the response
# Then format the response and give it as the output

chain = prompt | llm | StrOutputParser()
```

```python
topics = ['RAG', 'Agentic AI', 'Healthcare data engineer jobs']

for t in topics:
  print(f'\n Explaning: {t}')
  result = chain.invoke({'question': t})
  print(f'\n Response {result}')
```

```python
documents = ['we offer 30-day returns with receipts',
             'Free shipping on orders over $50',
             'support available 9 am - 5pm Monday-Friday',
             'We accept all major credit cards']
```

```python
from langchain.embeddings import HuggingFaceBgeEmbeddings
embeddings = HuggingFaceBgeEmbeddings(model_name = 'BAAI/bge-small-en-v1.5')
```

```python
from langchain.vectorstores import FAISS
from langchain.schema import Document
```

```python
documents
```

```python
docs = [Document(page_content = doc) for doc in documents]

docs
```

```python
vectorstore = FAISS.from_documents(docs, embeddings)
```

```python
vectorstore
```

```python
test_query = 'refund policy'
relavent_document= vectorstore.similarity_search(test_query, k=1)
print(f'{relavent_document[0].page_content}')
```

```python
print(relavent_document)
'\n'.join([doc.page_content for doc in relavent_document])
```

```python
def ask_rag(question):
  relavent_document = vectorstore.similarity_search(question, k=2)
  context = '\n'.join([doc.page_content for doc in relavent_document])
  rag_prompt = f"""
  Based on this company information:
  {context}

  Question: {question}
  please provide helpful answer based on the comapany information.
  Answer: """

  response = llm.invoke(rag_prompt)
  return response

question = 'what is your return policy'

answer = ask_rag(question)
print(f'\n Question: {question}')
print(f'\n Answer: {answer}')
```
